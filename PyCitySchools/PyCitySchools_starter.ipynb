```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = school_data['school_name'].nunique()

# Calculate the total number of students
total_students = student_data['Student ID'].nunique()

# Calculate the total budget (assuming each school has the same budget)
total_budget = school_data['budget'].sum()

# Calculate the average math and reading scores
average_math_score = student_data['maths_score'].mean()
average_reading_score = student_data['reading_score'].mean()

# Calculate the percentage passing math (score of 50 or greater)
passing_math_count = student_data[student_data['maths_score'] >= 50]['Student ID'].count()
percentage_passing_math = (passing_math_count / total_students) * 100

# Calculate the percentage passing reading (score of 50 or greater)
passing_reading_count = student_data[student_data['reading_score'] >= 50]['Student ID'].count()
percentage_passing_reading = (passing_reading_count / total_students) * 100

# Calculate the percentage overall passing (both math and reading)
overall_passing_count = student_data[(student_data['maths_score'] >= 50) & (student_data['reading_score'] >= 50)]['Student ID'].count()
percentage_passing_overall = (overall_passing_count / total_students) * 100

# Create a DataFrame to hold the above results
lga_summary_df = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [percentage_passing_math],
    '% Passing Reading': [percentage_passing_reading],
    '% Overall Passing': [percentage_passing_overall]
})

# Format the displayed data
lga_summary_df['Total Students'] = lga_summary_df['Total Students'].map("{:,}".format)
lga_summary_df['Total Budget'] = lga_summary_df['Total Budget'].map("${:,.2f}".format)
lga_summary_df['Average Math Score'] = lga_summary_df['Average Math Score'].map("{:.2f}".format)
lga_summary_df['Average Reading Score'] = lga_summary_df['Average Reading Score'].map("{:.2f}".format)
lga_summary_df['% Passing Math'] = lga_summary_df['% Passing Math'].map("{:.2f}%".format)
lga_summary_df['% Passing Reading'] = lga_summary_df['% Passing Reading'].map("{:.2f}%".format)
lga_summary_df['% Overall Passing'] = lga_summary_df['% Overall Passing'].map("{:.2f}%".format)

# Display the LGA summary DataFrame
lga_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group the data by school name to perform calculations for each school
school_grouped = school_data_complete.groupby('school_name')

# Calculate total students for each school
total_students_per_school = school_grouped['Student ID'].count()

# Calculate total school budget for each school
total_budget_per_school = school_grouped['budget'].max()

# Calculate per student budget for each school
per_student_budget_per_school = total_budget_per_school / total_students_per_school

# Calculate average math and reading scores for each school
average_math_score_per_school = school_grouped['maths_score'].mean()
average_reading_score_per_school = school_grouped['reading_score'].mean()

# Calculate the percentage passing math for each school
passing_math_per_school = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('school_name')['Student ID'].count()
percentage_passing_math_per_school = (passing_math_per_school / total_students_per_school) * 100

# Calculate the percentage passing reading for each school
passing_reading_per_school = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('school_name')['Student ID'].count()
percentage_passing_reading_per_school = (passing_reading_per_school / total_students_per_school) * 100

# Calculate the percentage overall passing for each school (both math and reading)
passing_overall_per_school = school_data_complete[(school_data_complete['maths_score'] >= 50) & (school_data_complete['reading_score'] >= 50)].groupby('school_name')['Student ID'].count()
percentage_passing_overall_per_school = (passing_overall_per_school / total_students_per_school) * 100

# Get the school type for each school
school_type_per_school = school_data.set_index('school_name')['type']

# Create a DataFrame to hold the above results
school_summary_df = pd.DataFrame({
    'School Type': school_type_per_school,
    'Total Students': total_students_per_school,
    'Total School Budget': total_budget_per_school,
    'Per Student Budget': per_student_budget_per_school,
    'Average Math Score': average_math_score_per_school,
    'Average Reading Score': average_reading_score_per_school,
    '% Passing Math': percentage_passing_math_per_school,
    '% Passing Reading': percentage_passing_reading_per_school,
    '% Overall Passing': percentage_passing_overall_per_school
})

# Format the displayed data
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map("${:,.2f}".format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map("${:,.2f}".format)

# Display the school summary DataFrame
school_summary_df
```

## Maths Scores by Year

  * Combine the series into a dataframe
